```python
import matplotlib.pyplot as plt
from colombia_hydrodata import Client

client = Client()
station = client.fetch_station("29037020")
dataset = station["NIVEL@NV_MEDIA_D"].sight_level(-0.367).rescale(1 / 100)
print(dataset)
```

```python
fig, axs = dataset.interpolate("ME").deconstruction().plot.time_series_analysis(figsize=(8, 4), tight_layout=True)
plt.show()
```

```python
fig, axs = dataset.interpolate("D").deconstruction().plot.daily_series_analysis(years=[2026, 2010], figsize=(8, 4), tight_layout=True)
axs[0].legend(title="Years")
plt.show()
```
